Question 1

Using the method of moments, we find the first moment $E(X)$

$$E(x) = \beta / (\beta - 1)$$


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/Sample data for Problem 1 Assignment 11 Fall 2025.xlsx", header=None)


# convert to NumPy array
x = df.values.flatten()      # 1D NumPy array
```

```python
xbar = np.mean(x)
print(xbar)
```

```python
beta_mom = xbar / (xbar - 1)

print("Sample mean:", xbar)
print("MoM estimate beta:", beta_mom)
```
